```python
import pandas as pd

file_path = 'P_CDQ.xpt'
file_path2 = 'P_DEMO.xpt'
df_cdq = pd.read_sas(file_path)
df_demo = pd.read_sas(file_path2)
df_cdq.head()
df_demo.head()
```

```python

df = pd.merge(df_demo, df_cdq, on='SEQN', how='inner')

print(df.head())
```

```python
# Create a heart condition indicator based on relevant columns
df['Heart_Issues'] = (
    (df['CDQ001'] == 1) |  # Chest pain present
    (df['CDQ002'] == 1) |  # Chest pain on exertion
    (df['CDQ010'] == 1)    # Shortness of breath
).astype(int)

# Check the distribution of the heart condition indicator
print(df['Heart_Issues'].value_counts())
```

```python
# Define the age groups
bins = [0, 20, 30, 40, 50, 60, 70, 100]
labels = ['10-20', '21-30', '31-40', '41-50', '51-60', '61-70', '70+']
df['Age_Group'] = pd.cut(df['RIDAGEYR'], bins=bins, labels=labels, right=False)

# Check the age groups
print(df['Age_Group'].value_counts())
```

```python
# Count the number of males and females in the study
gender_counts = df['RIAGENDR'].value_counts()
print("Gender distribution:")
print(gender_counts)

# Count the number of males and females with heart conditions
heart_gender_counts = df[df['Heart_Issues'] == 1]['RIAGENDR'].value_counts()
print("\nHeart condition by gender:")
print(heart_gender_counts)
```

```python
# Group by age and calculate the number of people with heart conditions
age_heart_counts = df[df['Heart_Issues'] == 1].groupby('Age_Group').size()
print("\nHeart condition by age group:")
print(age_heart_counts)
```

```python
# Calculate the total number of people in each age group
age_total_counts = df.groupby('Age_Group').size()

# Calculate the percentage of heart conditions in each age group
age_percentage = (age_heart_counts / age_total_counts) * 100
print("\nPercentage of heart conditions by age group:")
print(age_percentage)
```

```python
# Summary results
summary = {
    'Total People with Heart Conditions': df[df['Heart_Issues'] == 1].shape[0],
    'Gender Distribution': gender_counts,
    'Heart Condition by Gender': heart_gender_counts,
    'Heart Condition by Age Group': age_heart_counts,
    'Heart Condition Percentage by Age Group': age_percentage
}

print("\nSummary of Results:")
print(summary)
```
